```python
time_delta = pd.to_datetime(df['DD_DELIVERY_TIME']) - pd.to_datetime(df['EX_START_TIME'] )
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.clf()

time_del_hours = (time_delta/ np.timedelta64(1, 'h'))
df['admin_time_delta'] = time_del_hours
#df[['admin_time_delta', 'DD_DELIVERY_METHOD']].groupby('DD_DELIVERY_METHOD').plot(kind='kde')
sns.kdeplot(df['admin_time_delta'][df['DD_DELIVERY_METHOD'] == 'CESAREAN SECTION'], label='CESAREAN SECTION', shade=True)
sns.kdeplot(df['admin_time_delta'][df['DD_DELIVERY_METHOD'].isin(['INSTRUMENTED VAGINAL', 'SPONTANEOUS VAGINAL'])], label='VAGINAL', shade=True)
ax = sns.kdeplot(df['admin_time_delta'], label='COMBINED', shade=True)
plt.xlabel('Time Between Propess Administration and Delivery [Hours]')
plt.ylabel('Density')
sns.plt.xlim(0,300)
ax.xaxis.set_major_locator(ticker.MultipleLocator(20))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
#plt.show()
plt.savefig('F:/Projects/Ferring/results/pre_modelling/administration_delivery_time_delta_del_type.png')
```

```python
plt.clf()
time_del_hours[time_del_hours<100].plot(kind='hist', bins=20)
plt.xlabel('Time Between Propess Administration and Delivery [Hours]')
plt.ylabel('Number of Patients')
#plt.show()
plt.savefig('F:/Projects/Ferring/results/pre_modelling/administration_delivery_time_delta_lt_100.png')
```
